```python
# Inicializar o detector MTCNN
detector = MTCNN()
```

```python
# Carregar imagem
img = cv2.imread('./imagem_teste.jpg')
```

```python
# Detectar rostos
faces = detector.detect_faces(img)
```

```python
# Desenhar caixas delimitadoras
for face in faces:
    x, y, w, h = face['box']
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
```

```python
# Exibir imagem
print('Rostos detectados')
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
